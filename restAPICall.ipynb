```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = "https://www.gov.uk/government/publications/renewable-energy-planning-database-monthly-extract"

response = requests.get(base_url)
soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all("a")

csv_links = []

for link in links:
    href = link.get("href")
    if href and ".csv" in href.lower():
        full_url = urljoin(base_url, href)
        csv_links.append(full_url)

csv_links
```

```python
import re

def extract_date(url):
    match = re.search(r'(\d{4})_Q(\d)', url)
    if match:
        year = int(match.group(1))
        quarter = int(match.group(2))
        return (year, quarter)
    return (0, 0)

latest_csv = max(csv_links, key=extract_date)

latest_csv
```
